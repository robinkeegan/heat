```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
x = np.linspace(1,10,10)
y = x
```

```python
plt.figure(figsize =(8,6))
plt.plot(x,y, c= 'k', marker ='o', label = 'label')
plt.xlabel('xlab')
plt.ylabel('ylab')
plt.show()
```

```python
def plot(x,y, xlab = '', ylab ='', c = 'k', figsize = (6,4), marker = True, save = False, fname = None):
    plt.figure(figsize =figsize)
    #Marker or not
    if marker == True:
        plt.plot(x,y, c= c, marker ='o')
    elif marker == False:
        plt.plot(x,y, c= c)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    #Show or save figure
    if save == True:
        if fname == None:
            print("Error is saving a filename must be specified")
        else:
            plt.tight_layout()
            plt.savefig(fname)
    elif save == False:
        plt.show()
```

```python
plot(x,y, c = 'r', figsize = (3,8), save = True, fname = None)
```
